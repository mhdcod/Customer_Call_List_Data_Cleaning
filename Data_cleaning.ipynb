```python
import pandas as pd
```

# data cleaning of a to be called customer data set

```python
df = pd.read_excel("/Users/mohamed/Desktop/P/data_cleaning_Prj1/Original_Customer_List .xlsx")
df.head(25)
```

```python
# we can see there are few things to clean : caracter in Last_Name column, different  Phone_Number format, some adresses have a zip code others don't, same lines, Nan, Yes Y No N, unuseful last column 
```

```python
df = df.drop_duplicates()
df
#got rid of the duplicate line
```

```python
df = df.drop(columns = 'Not_Useful_Column')
df 
#got rid of the unecessary column 
```

```python
df['Last_Name'] = df['Last_Name'].str.lstrip('...')
df
#got rid of the '...' before Last_Name Potter row 11
```

```python
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
df['Last_Name'] = df['Last_Name'].str.rstrip('_')
df 
#got rid of typos at the beg&end of LastNames
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True)
df
#got rid of any caracter not a number or a letter : standardize our Phone numbers 
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x : str(x))
df
# converted our phone numbers to string 
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x : x[0:3]+'-'+x[3:6]+'-'+x[6:10])
df
#inserted '-' to standardize phone numbers 
```

```python
df["Phone_Number"] = df["Phone_Number"].replace('Na--','')
df["Phone_Number"] = df["Phone_Number"].replace('nan--','')
df 
#got rid of the 'Na/nan' values by remplacing them with a blank
```

```python
df['Address'].str.split(',',n=2, expand=True)
#splited street state and zipcode
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',',n=2, expand=True)
#created new columns with the splited values of Adress column 
```

```python
df
```

```python
df = df.drop(columns = 'Not_Useful_Column')
df
#had to drop the unuseful once again don't know why it showed up again 
```

```python
df['Paying Customer'].str.replace('Y','Yes')
#gotta do the other way around yes to y + less data u have better it is 
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df.head()
#standardized the paying customer column 
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df.head(10)
#standardized the Do_Not_Contact column 
```

```python
df = df.fillna('')
df
#inserted a blanked value into all fields that hadnot NaN(int) or None(str)
```

```python
df = df.replace('N/a','')
df 
#got rid of the N/a which is actually a value unlike NaN
```

```python
for x in df.index: 
    if df.loc[x,'Phone_Number'] == '':
        df.drop(x, inplace=True)

df
#created a loup to drop all blankes rows of PhoneNumber column 
```

```python
for x in df.index: 
    if df.loc[x,'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)

df
#created a loup to drop all Yes rows of Do_Not_Contact column 
```

```python
df = df.reset_index(drop=True)
#reset index 
```

```python
df
```

```python
df = df.drop(columns ='Zip_Code')
df
#droped zipcode column as its empty 
```

```python
df = df[['CustomerID', 'First_Name', 'Last_Name', 'Phone_Number', 'Street_Address', 'State', 'Address', 'Paying Customer', 'Do_Not_Contact']]
df = df.drop(columns='Address')

#reorganized order of columns and droped unuseful ones 
```

```python
df
```

```python
df.to_excel('cleaned_data.xlsx', index=False)
```
